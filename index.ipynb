```python

from transformers import ViTFeatureExtractor, ViTForImageClassification
from PIL import Image
import requests
import torch
import matplotlib.pyplot as plt
```

```python
feature = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')

model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')
```

```python
image = Image.open('Image.jpg')

plt.imshow(image)
```

```python
input = feature(images=image, return_tensors="pt")

pixle = input['pixel_values']

output = model(pixle)
```

```python
logits = output.logits
```

```python
predict = logits.argmax(-1).item()

predict
```

```python
predict = model.config.id2label[predict]

predict
```
